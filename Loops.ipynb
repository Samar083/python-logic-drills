## Top 5 interview question on loops

##### 1. The "FizzBuzz" (The Universal Entry Question)
###### The Goal: Iterate from 1 to 50.

###### If a number is divisible by 3, print "Fizz".

###### If divisible by 5, print "Buzz".

###### If divisible by both 3 and 5, print "FizzBuzz".

##### Otherwise, print the number.

```python
for i in range(1,51):
    if (i % 3 == 0) and (i % 5 == 0):
        print("FizzBuzz")
    elif (i % 3 == 0):
        print("Fizz")
    elif (i % 5 == 0):
        print("Buzz")
else:
    print("done")
    
```

## 2. Reverse a String (Without using [::-1])

```python
a = input("Enter you string: ")
b = ""
for i in a:
    b = i + b ## prepending method
print(b)
    
```

## 3. Finding the "Peak" (Max Value)
#### The Goal: Given a list nums = [12, 45, 2, 89, 34], find the largest number without using the built-in max() function.

```python
nums = [12, 45, 2, 89, 34]
highest = nums[0]
for i in nums:
   if i > highest:
       highest = i
print(highest)  
```

```python
nums = [12, 45, 2, 89, 34]
lowest = nums[0]
for i in nums:
    if i < lowest:
        lowest = i
print(lowest)
    
```

#### Removing Duplicates from a List
###### The Goal: Given data = [1, 2, 2, 3, 4, 4, 4, 5], create a new list that contains each number only once: [1, 2, 3, 4, 5].

```python
data = [1, 2, 2, 3, 4, 4, 4, 5]
new_list = []
for i in data:
    if i not in new_list:
        new_list.append(i)
print(new_list)
```
